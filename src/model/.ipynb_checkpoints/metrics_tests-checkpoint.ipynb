```python
# Setup and Imports
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

from metrics_clean_data import collect_cleaned_lyrics
from metrics import (
    average_line_length,
    song_word_variation,
    genre_word_variation,
    i_vs_you_score,
    count_word_repetitions
)
```

```python
# Load pre-cleaned songs and genre labels from CSV files
songs, genres = collect_cleaned_lyrics("../../data")

# Print sample information
print("Sample Lyrics Snippet:\n", songs[0][:300])
print("Genres Found:", set(genres))
print("Total Songs loaded:", len(songs))
```

```python
# Group lyrics by genre
genre_data = defaultdict(list)
for lyric, genre in zip(songs, genres):
    genre_data[genre].append(lyric)

# Show how many songs per genre
for genre, lyrics in genre_data.items():
    print(f"{genre}: {len(lyrics)} songs")
```

```python
# Compute metrics for each individual song
results = []
for genre, lyric_list in genre_data.items():
    for lyrics in lyric_list:
        results.append({
            "genre": genre,
            "avg_line_length": average_line_length(lyrics),
            "word_variation": song_word_variation(lyrics),
            "i_vs_you": i_vs_you_score(lyrics),
            "repetitions": count_word_repetitions(lyrics)
        })

df = pd.DataFrame(results)
df.head(10)
```

```python
# Display the first 5 rows of metrics for the first 5 genres
first_genres = list(df["genre"].unique())[:5]  # grab first 5 unique genres

for genre in first_genres:
    print(f"Top 5 Songs in Genre: {genre.upper()}")
    display(df[df["genre"] == genre].head(5))  # use display for clean output in Jupyter
    print("\n" + "-"*60 + "\n")
```

```python
# Generic plotting function
def plot_metric_by_genre(df, metric_name, ylabel):
    grouped = df.groupby("genre")[metric_name].mean().sort_values()
    plt.figure(figsize=(14, 7))
    grouped.plot(kind="bar", color="cornflowerblue", edgecolor="black")
    plt.title(f"Average {metric_name.replace('_', ' ').title()} by Genre")
    plt.ylabel(ylabel)
    plt.xlabel("Genre")
    plt.xticks(rotation=60, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
# Visualize each metric
plot_metric_by_genre(df, "avg_line_length", "Avg Words per Line")
plot_metric_by_genre(df, "word_variation", "Unique/Total Words")
plot_metric_by_genre(df, "i_vs_you", "I vs. You (Positive = More I)")
plot_metric_by_genre(df, "repetitions", "Repetitions per Song")
```
